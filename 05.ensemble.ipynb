```python
import pandas as pd
import os
```

# Data preparation

```python
resPath = 'result'
```

```python
files = os.listdir(resPath)
```

```python
files = [os.path.join(resPath, i) for i in files]
```

```python
df=pd.read_csv(files[0], index_col=0)
```

```python
df2=pd.read_csv(files[1], index_col=0)
```

```python
df3=pd.read_csv(files[2], index_col=0)
```

```python
df.columns = ['日期', '時間', 'if']
```

```python
df['lof']=df2['outlier']
```

```python
df['md']=df3['outlier']
```

# Anomaly rankings

```python
anomaly = df.apply(lambda x: sum(x[['lof','md','if']]), axis=1)
```

```python
df['anomaly']=anomaly
```

```python
switch_dict = {'0':'green', '1':'yellow', '2':'orange', '3':'red'}
df['rank']=df.apply(lambda x: switch_dict.get(str(x['anomaly'])), axis=1)
```

```python
df[df['rank']=='yellow'][['日期','時間', 'rank']]
```
